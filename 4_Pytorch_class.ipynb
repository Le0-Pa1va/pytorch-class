# CAMADA LINEAR


```python
import torch
from torch import nn # neural network
```

```python
torch.manual_seed(42)

# perceptron tem x1, x2, x3 e retorna apenas um resultado
perceptron = nn.Linear(in_features=3, out_features=1)
print(perceptron)
```

```python
for name, tensor in perceptron.named_parameters():
  print(name, '\t', tensor.data)

print('')
print(perceptron.weight.data)
print(perceptron.bias.data)
```

```python
w1, w2, w3 = perceptron.weight.data.numpy()[0]
b = perceptron.bias.data.numpy()[0]
# w1 * x1 + w2 * x2 + w3 * x3 + b = 0

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def plot3d(perceptron):
  w1, w2, w3 = perceptron.weight.data.numpy()[0]
  b = perceptron.bias.data.numpy()[0]

  X1 = np.linspace(-1, 1, 10)
  X2 = np.linspace(-1, 1, 10)

  X1, X2 = np.meshgrid(X1, X2)
  X3 = (b - w1*X1 + w2*X2) / w3

  fig = plt.figure(figsize=(10, 8))
  ax = fig.add_subplot(projection='3d')
  ax.view_init(azim=180)
  ax.plot_surface(X1, X2, X3, cmap='plasma')

plot3d(perceptron)
```

```python
X = torch.Tensor([0, 1, 2])
y = perceptron(X)
print(y)
plot3d(perceptron)

plt.plot([X[0]], [X[1]], [X[2]], color='r', marker='^', markersize=20)
```
